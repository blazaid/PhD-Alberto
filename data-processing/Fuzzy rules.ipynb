```python
antecedents = []
for variable, fss in variables:
    var_antecedents = [f'{variable} IS {fs}' for fs in fss]
    antecedents.append(var_antecedents)

rules = []
for output in (o1, o2):
    for ants, var in zip(itertools.product(*antecedents), output[1]):
        rule = list(ants) + [output[0], np.around(1 / (1 + math.exp(-var)), 2)]
        rules.append(rule)
```

## Simplificamos

Primero voy a eliminar aquellas reglas que son exactamente iguales en el antecedente que activan ambos consecuentes

```python
mid = len(rules) // 2
sorted_rules = sorted(rules, key=lambda x: (x[-2], x[:-2]))
rules_o1, rules_o2 = sorted_rules[:mid], sorted_rules[mid:]
relevant_rules = []
for r1, r2 in zip(rules_o1, rules_o2):
    #if np.isclose(r1[-1], r2[-1]):
    #    print(r1)
    #    print(r2)
    #    print('-------------------------------')
    if not np.isclose(r1[-1], r2[-1]):
        relevant_rules.append(r1)
        relevant_rules.append(r2)
len(relevant_rules)
```

Ahora eliminamos aquellas que tienen una activación muy baja (por debajo de 0.01)

```python
relevant_rules = [r for r in relevant_rules if r[-1] > 0.01]
len(relevant_rules)
```

```python
ants = antecedents[:]
remaining_rules = relevant_rules[:]

for i in range(len(antecedents)):
    simplified_rules = []
    while remaining_rules:
        rule1 = remaining_rules.pop(0)
        rule1_components = [x for n, x in enumerate(rule1[:-1]) if n != i]
        rule1_weight = rule1[-1]
        
        rule_found = False
        for n, rule2 in enumerate(simplified_rules):
            rule2_components = [x for k, x in enumerate(rule2[:-1]) if k != i]
            rule2_weight = rule2[-1]
            if rule1_components[:-1] == rule2_components[:-1] and np.isclose(rule1_weight, rule2_weight, 0.01, 0.01):
                simplified_rules[n] = rule2_components + [rule2_weight]
                rule_found = True
        if not rule_found:
            simplified_rules.append(rule1)
        
    remaining_rules = simplified_rules
len(simplified_rules)
```

```python
with open('rules.fz', 'w') as f:
    for rule in simplified_rules:
        f.write(f'IF {" AND ".join(rule[:-2])} THEN {rule[-2]} WITH {rule[-1]}')
        f.write('\n')
```
